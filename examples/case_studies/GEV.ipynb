# Generalized Extreme Value Distribution

The Generalized Extreme Value (GEV) distribution is a meta-distribution containing the Weibull, Gumbel, and Frechet families of extreme value distributions. It is used for modelling the distribution of extremes (maxima or minima) of stationary processes, such as the annual maximum wind speed, annual maximum truck weight on a bridge, and so on.

Following the parametrization used in Coles (2001), it is given by:

$$G(x) = \exp \left\{ \left[ 1 - \xi \left( \frac{x - \mu}{\sigma} \right) \right]^{-\frac{1}{\xi}} \right\}$$

when:
- $\xi < 0$ we get the Weibull distribution with a bounded upper tail;
- $\xi = 0$, in the limit, we get the Gumbel distribution, unbonded in both tails;
- $\xi > 0$, we get the Frechet distribution which is bounded in the lower tail.

Note that this parametrization of the shape parameter $\xi$ is opposite to that used in SciPy.

We will use the example of the Port Pire annual maximum sea-level data used in Coles (2001), and compare with the frequentist results presented there.

- Coles, S. (2001), *An Introduction to the Statistical Modeling of Extreme Values*, Springer-Verlag, London

## Modelling & Prediction
In the modelling we wish to do two thing:

- parameter inference on the GEV parameters, based on some fairly non-informative priors
- prediction of the 10-year return level

Predictions of extreme values considering parameter uncertainty are easily accomplished in the Bayesian setting. It is interesting to compare this ease to the difficulties encountered by [Caprani & OBrien (2010)](https://doi.org/10.1016/j.strusafe.2009.09.001) in doing this in a frequentist setting. In any case, the predicted value at a probability of exceedance $p$ is given by:

$$ x_p = \mu - \frac{\sigma}{\xi} \left\{1 - \left[-\log\left(1-p\right)\right] \right\} $$

This is a deterministic function of the parameter values, and so is accomplished using `pm.Deterministic` within the model context.

Consider then, the 10-year return period, for which $p = 1/10$:

```python
p = 1/10
```

And now set up the model using priors estimated from a quick review of the historgram above:

- $\mu$: there is no real absis for considering anything other than a `Normal` distribution;
- $\sigma$: this must be positive, and has a small value, so use `HalfNormal`
- $\xi$: we are agnostic to the tail behaviour so centre this at zero, but limit to physically reasonable bounds of $\pm 0.6$

```python
# Optionally centre the data, depending on fitting and divergences
# cdata = (data - data.mean())/data.std()

with pm.Model() as model:
    # Priors
    μ = pm.Normal('μ', mu=3.8, sigma=0.2)
    σ = pm.HalfNormal('σ', sigma=0.3)
    ξ = pm.TruncatedNormal('ξ', mu=0, sigma=0.2, lower=-0.6, upper=0.6)
    
    # Estimation
    gev = pm.GenExtreme('gev',mu=μ, sigma=σ, xi=ξ, observed=data)
    # Return level
    z_p = pm.Deterministic('z_p', μ - σ/ξ*(1 - (-np.log(1-p))**(-ξ)) )   
```

## Prior Predictive Checks
Let's get a feel for how well our selected priors cover the range of the data

```python
prior_pc = pm.sample_prior_predictive(samples=1000,model=model)
az.plot_ppc(prior_pc,group="prior",figsize=(12,6));
ax = plt.gca()
ax.set_xlim([2,6]);
ax.set_ylim([0,2]);
```

```python
data = prior_pc['prior']
fig,axs = plt.subplots(ncols=2,nrows=2,figsize=(12,12))
az.plot_dist(data.z_p,ax=axs[0,0],label='z_p')
az.plot_dist(data.μ,ax=axs[0,1],label='μ')
az.plot_dist(data.σ,ax=axs[1,0],label='σ')
az.plot_dist(data.ξ,ax=axs[1,1],label='ξ');
```

## Inference
Press the magic Inference Button$^{\mathrm{TM}}$

```python
with model:
    trace = pm.sample(5000, cores=4, chains=4, tune=2000, return_inferencedata=True, 
                      initvals={'μ': -0.5, 'σ': 1.0, 'ξ': -0.1}, target_accept=0.95)
```

```python
az.plot_trace(trace,var_names=['μ','σ','ξ'],figsize=(12,9));
```

The 95% CI range of the parameter estimates is:

```python
az.hdi(trace, hdi_prob=.95)
```

And examine the prediction distribution, considering parameter variability

```python
az.plot_posterior(trace,hdi_prob=0.95,var_names=['z_p'],round_to=4);
```

## Comparison
To compare with the results given in Coles (2001), we extract the maximum likelihood estimates (MLE) as the mode of the posterior distributions. 

The MLE results given in Coles (2001) are:

$$\left(\hat{\mu}, \hat{\sigma}, \hat{\xi} \right) = \left( 3.87, 0.198, -0.050 \right) $$


And the variance-covariance matrix of the estimates is:

$$ V = \left[ \begin{array} 0.000780 & 0.000197 & -0.00107 \\
                            0.000197 & 0.000410 &  -0.000778 \\
                            -0.00107 & -0.000778 & 0.00965 
              \end{array} \right] $$


Note that extracting the MLE estimates from our inference involves accessing some of the Arviz back end functions to bash the xarray into something examinable:

```python
_,vals = az.sel_utils.xarray_to_ndarray(trace['posterior'],var_names=['μ', 'σ', 'ξ'])
mle = [azpu.calculate_point_estimate('mode',val) for val in vals]
mle
```
